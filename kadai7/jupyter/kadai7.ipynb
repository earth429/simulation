```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("kadai7-4.csv",header=None)
df = df.rename(columns={0:"step",1:"y",2:"v"})
df.head()
```

```python
plt.figure(figsize=(6,4),dpi=200,facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(df["step"],df["y"],label="y",linewidth=3,marker="x")
plt.plot(df["step"],df["v"],label="v",linewidth=3,marker="o")
plt.xlabel("$t$",fontsize=15)
plt.ylabel("$u$",fontsize=15)
plt.legend()
plt.savefig("kadai7-4.pdf")
```
