```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./RandomWalk.csv')
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('$N$')
ax.set_ylabel('$V$')
ax.set_title("Random Walk")
ax.grid()

x = df['n']
a1 = -542.382933
a2 = 1.054153
y = a2 * x + a1
ax.plot(df['n'], df['v'], marker="o", label="$V$")
ax.plot(x, y, marker="x", label="line")
a = 120.753067
b = 0.391017
c = 0.000121
y = c * x ** 2 + b * x + a
ax.plot(x, y, marker="s", label="quadratic")
ax.legend(loc=0)    
fig.tight_layout()
plt.savefig("RandomWalk.pdf",bbox_inches='tight')
plt.show()
```
