```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
df = pd.read_csv("kadai10-2.csv",header=None)
df = df.rename(columns={0:"step",1:"X",2:"Vx",3:"Y",4:"Vy",5:"Z",6:"Vz"})
df.head()
```

```python
# 3Dでプロット
fig = plt.figure(facecolor="white",dpi=100)
ax = Axes3D(fig)
ax.plot(df["X"], df["Y"], df["Z"], "o-", color="#0000aa", ms=2, mew=0.5,clip_on=False)

# 軸ラベル
ax.set_xlabel('x',fontsize=15)
ax.set_ylabel('y',fontsize=15)
ax.set_zlabel('z',fontsize=15)
#plt.savefig("xyz.png",bbox_inches='tight')
plt.savefig("kadai10-2.pdf",bbox_inches='tight')
# 表示
plt.show()
```
