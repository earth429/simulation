```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("kadai6-4.csv",header=None)
df = df.rename(columns={0:"step",1:"y1",2:"y2"})
df.head()
```

```python
plt.figure(figsize=(6,4),dpi=200,facecolor="white")
plt.tick_params(direction='in',#
                pad =10, #
                length=7.5, #
                width=1.5, #
                labelsize=12,
               )
plt.plot(df["step"],df["y1"],label="$y_1$",linewidth=3,marker="x")
plt.plot(df["step"],df["y2"],label="$y_2$",linewidth=3,marker="o")
plt.xlabel("$t$",fontsize=15)
plt.ylabel("$y$",fontsize=15)
plt.legend()
plt.savefig("kadai6-4.pdf")
```
