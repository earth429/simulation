```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
h1 = pd.read_csv("./csv/Euler01.csv",header=None, names=['x', 'y'])
#h1 値が確認できる
```

```python
h2 = pd.read_csv("./csv/Euler02.csv",header=None, names=['x', 'y'])
#h2
```

```python
h05 = pd.read_csv("./csv/Euler005.csv",header=None, names=['x', 'y'])
#h05
```

```python
h5 = pd.read_csv("./csv/Euler05.csv",header=None, names=['x', 'y'])
#h5
```

```python
truth_x = np.linspace(-0.5,1.5,1000)
truth_y = np.e**(truth_x)
plt.figure(facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(truth_x,truth_y,label="u = e^t",linewidth = 3.0)
plt.scatter(h05["x"],h05["y"],label="h= 0.05",marker="x",s=100)
plt.scatter(h1["x"],h1["y"],label="h= 0.1",s=100)
plt.scatter(h2["x"],h2["y"],label="h= 0.2",marker="s",s=100)
plt.scatter(h5["x"],h5["y"],label="h= 0.5",marker="^",s=100)
plt.xlabel("t",fontsize=13)
plt.ylabel("u",fontsize=13)
plt.xlim(-0.1,1.1)
plt.ylim(0.8,3)
plt.legend()
plt.title("Euler method")
plt.savefig("Euler.pdf")
```
